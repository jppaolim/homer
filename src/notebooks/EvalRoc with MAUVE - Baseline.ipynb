```python
NUM_SAMPLES_X = 10000

BaseRepoPath = "/home/studio-lab-user/homer22/homer-22/"
dataPath=BaseRepoPath+"data/"
RunName="r2"

import transformers
```

```python
from datasets import load_metric
metric = load_metric('mauve')
```

```python
from datasets import load_from_disk
ds = load_from_disk(dataPath+"TitleAndStory/")
ds['train']
```

```python
import random
index = random.sample(range(ds['train'].num_rows),NUM_SAMPLES_X)

middle_index = int(NUM_SAMPLES_X / 2)
first_half=index[:middle_index]
second_half=index[middle_index:]
```

```python
HumanStories=ds['train'].select(first_half)["fullstory"]
OtherHumanStories=ds['train'].select(second_half)["fullstory"]
```

```python
import time 
StartTime = time.time()
print(StartTime)
```

```python
out = metric.compute(predictions=HumanStories, references=OtherHumanStories)
print(out.mauve)
```

```python
MauvTime = time.time() -StartTime
```

```python
import json
result = { 'model' : "Human", 'Run' : RunName, 'Samples' :  middle_index,  'mauve' : out.mauve, 'Gentime' : "NA", 'MauvTime' : MauvTime }
```

```python
result
```

```python
try:
  with open('EvalResult.json') as f:
    data = json.load(f)
except FileNotFoundError:
  data = {'results' : []}

data['results'].append(result)

with open('EvalResult.json', 'w') as f:
    json.dump(data, f)
```
