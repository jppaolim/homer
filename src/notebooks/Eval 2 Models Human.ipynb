```python
import random 
import csv
import json
import torch
import os
```

```python
ModelAName="v48_GPT2Medium_PT"
RunName="r1"

ModelBName="v56_Large_2E"
RunName="r1"

#BaseRepoPath = "/home/studio-lab-user/homer22/homer-22/"
#BaseRepoPath = "/Users/jppm/Coding/MyCode/homer-22/"

ISGPU = torch.cuda.is_available()
device = "cuda:0" if ISGPU else "cpu"


random.seed(42)
```

```python
ModelA_File="./Stories/"+ModelAName+RunName+"MachineStories.csv"
ModelB_File="./Stories/"+ModelBName+RunName+"MachineStories.csv"

ModelA_Stories = []
ModelB_Stories  = []

with open(ModelA_File, 'r') as csvfile:
    reader = csv.reader(csvfile, quoting=csv.QUOTE_ALL)
    for row in reader:
        ModelA_Stories.append(row[0])

with open(ModelB_File, 'r') as csvfile:
    reader = csv.reader(csvfile, quoting=csv.QUOTE_ALL)
    for row in reader:
        ModelB_Stories.append(row[0])

if len(ModelA_Stories) != len(ModelB_Stories):
    print("Warning, 2 different lists ")
    exit

NUM_SAMPLES = len(ModelA_Stories)
```

```python
Total_Samples = 0 

lst = list(range(0,NUM_SAMPLES))
random.shuffle(lst)

Stories=[[ModelA_Stories,0],[ModelB_Stories, 0]] 
```

```python
from IPython.display import clear_output


try:
    while True:
        clear_output(wait=True)

        #AorB = random.randint(0,1)
        AorB = 0
        i = lst[Total_Samples]

        print("Sampling story : {} ".format(i))
        print("Model 1 :")  
        print(Stories[AorB][0][i])
        
        print("Model 2 :")  
        print(Stories[1-AorB][0][i])

        d1a = input ("Do you want prefer : A) Story 1  B) Story 2 or Q to interrupt [A/B/Q]? : ")
        if d1a == "A":
            Stories[AorB][1] += 1
        elif d1a == "B":
            Stories[1-AorB][1] += 1
        elif d1a == "Q":
            break
        #Besoin de revoir ça .... 
        else:
            break

        Total_Samples += 1

except KeyboardInterrupt:
    print('interrupted!')

```

```python
print("total Number reviewed : {} ".format(Total_Samples))

APref = Stories[0][1]
BPref = Stories[1][1] 

print("total Pref model {} : {} so {:.1f}%".format(ModelAName, APref, APref/Total_Samples * 100))
print("total Pref model {} : {} so {:.1f}%".format(ModelBName, BPref, BPref/Total_Samples * 100))
```
